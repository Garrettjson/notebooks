#### Numpy Array Slicing

It's possible to slice numpy arrays. Since this is a 3D array, the first number decides which 2D subarray block to select, the second number decides which rows in the selected 2D subarray block to select, and the third number selects which columns in the 2D subarray block. Using the `:` operator selects everything.

Here, the first 3 columns in each subarry represent a color channel (RGB), with the last channel representing brightness. Selecting just one column will yield an interesting result

```python
x = img[:,:,0]
plt.imshow(x, interpolation='nearest')
```

Now selecting the first 75 rows (of pixels), and the all columns of pixels after index 100

```python
x = img[:75,100:,:]
plt.imshow(x, interpolation='nearest')
```
